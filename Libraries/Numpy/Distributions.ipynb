```python
from numpy import random as rd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

##  Normal (Gaussian) Distribution

NumPy provides the `normal()` function in the `random` module to generate data that follows a **normal distribution** (bell curve).

####  Syntax:
`numpy.random.normal(loc=0.0, scale=1.0, size=None)`
- `loc`: Mean (center) of the distribution.
- `scale`: Standard deviation (spread or width of the curve).
- `size`: Number of random values (shape of the output array).


```python
# Generate 1000 samples with mean = 0 and standard deviation = 1
data = rd.normal(loc=0.0, scale=3.0, size=1000)

# Plotting 
sns.kdeplot(data, fill=True, color='red')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

##  Binomial Distribution in NumPy

The **binomial distribution** is a discrete probability distribution that models the **NUMBER OF SUCCESSES IN FIXED NUMBER** of independent experiments, each with the same probability of success.
#### ðŸ§ª Syntax:
`numpy.random.binomial(n, p, size=None)`

- `n`: Number of trials.
- `p`: Probability of success on each trial.
- `size`: Number of experiments or shape of output array.

`Discrete Distribution`:The distribution is defined at separate set of events, e.g. a coin toss's result is discrete as it can be only head or tails whereas height of people is continuous as it can be 170, 170.1, 170.11 and so on.


```python
plt.hist(rd.binomial(n=100, p=0.5, size=100), color='lightgreen', edgecolor='black')
plt.title('Binomial Distribution Histogram (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
sns.kdeplot(rd.binomial(n=100, p=0.5, size=100), fill=True, color='blue')

plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
data = {
  "normal": rd.normal(loc=50, scale=5, size=1000),
  "binomial": rd.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

##  Poisson Distribution in NumPy

The **Poisson distribution** is used to model the **NUMBER OF TIME AN EVENT OCCURES IN A FIXED** interval of time or space. It assumes these events occur with a known constant rate and independently of the time since the last event.
### ðŸ§ª Syntax:
`numpy.random.poisson(lam=1.0, size=None)`

- `lam`: (Î») Expected number of events (mean rate of occurrence).
- `size`: Number of samples to generate (shape of output array).


```python
plt.hist(rd.poisson(lam= 5, size= 1000), color='lightblue', edgecolor='black', align='left')
plt.title('Poisson Distribution Histogram (Î» = 5)')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
data = {
  "normal": rd.normal(loc=50, scale=5, size=1000),
  "binomial": rd.binomial(n=100, p=0.5, size=1000),
  "poisson": rd.poisson(lam=50, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

## Uniform Distribution in NumPy

The **Uniform distribution** is used to generate numbers that are **evenly distributed** over a specified interval. Each value within the interval is equally likely to occur.
### Syntax:
`numpy.random.uniform(low=0.0, high=1.0, size=None)`
- `low`: Lower bound of the interval (inclusive). Default is 0.0.
- `high`: Upper bound of the interval (exclusive). Default is 1.0.
- size: Number of random values or the shape of the returned array.

```python
sns.kdeplot(rd.uniform(size=1000), fill=True, color='purple')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

##  Logistic Distribution in NumPy

The **Logistic Distribution** is similar to the normal distribution but has **sharper peaks** and **heavier tails**. It is often used in machine learning for logistic regression and in modeling population growth.
###  Syntax:
`numpy.random.logistic(loc=0.0, scale=1.0, size=None)`
- `loc`: Mean or center of the distribution.
- `scale`: Controls the spread (standard deviation â‰ˆ 1.814 Ã— scale).
- `size`: Number or shape of samples.


```python
import seaborn as sns
sns.kdeplot(rd.logistic(loc=0, scale=1, size=1000), fill=True, color='darkred')
plt.title('Logistic Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
data = {
  "normal": rd.normal(scale=2, size=1000),
  "logistic": rd.logistic(size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```
