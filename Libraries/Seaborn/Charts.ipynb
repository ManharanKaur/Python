```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'anime' : [
        "Cowboy Bebop",
        "Cowboy Bebop: Tengoku no Tobira",
        "Witch Hunter Robin",
        "Hachimitsu to Clover",
        "Hungry Heart: Wild Striker",
        "Initial D Fourth Stage",
        "Monster",
        "Naruto",
        "One Piece",
        "Tennis no Ouji-sama"
    ],
    'Score-1': [1580, 379, 131, 321, 32, 259, 1177, 5310, 6340, 225],
    'Score-2': [741, 109, 164, 278, 31, 48, 593, 3582, 2616, 210],
    'Score-3': [1357, 221, 353, 484, 88, 97, 882, 6098, 3675, 443]
}
df = pd.DataFrame(data)

# Melting the DataFrame
df_melted = pd.melt(df, id_vars=['anime'], var_name='Score Type', value_name='Score')

# Create the line plot
sns.lineplot(x='anime', y='Score', data=df_melted, hue='Score Type', marker='o')

plt.title('Sample Line Plot')
plt.xlabel('Anime')
plt.ylabel('Score')

plt.xticks(rotation=30, ha='right')
plt.show()
```
