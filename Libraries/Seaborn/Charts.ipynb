```python
# Dataset

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'anime' : [
        "Cowboy Bebop",
        "Cowboy Bebop: Tengoku no Tobira",
        "Witch Hunter Robin",
        "Hachimitsu to Clover",
        "Hungry Heart: Wild Striker",
        "Initial D Fourth Stage",
        "Monster",
        "Naruto",
        "One Piece",
        "Tennis no Ouji-sama"
    ],
    'Score-1': [1580, 379, 131, 321, 32, 259, 1177, 5310, 6340, 225],
    'Score-2': [741, 109, 164, 278, 31, 48, 593, 3582, 2616, 210],
    'Score-3': [1357, 221, 353, 484, 88, 97, 882, 6098, 3675, 443],
    'Completed': [718161, 208333, 46165, 81145, 13778, 90967, 214491, 1462223, 33, 76881],
    'Dropped': [26678, 770, 5378, 11026, 1168, 1356, 23008, 99806, 136245, 12516],
    'Watching': [105808, 4143, 4300, 11909, 817, 6082, 64648, 137167, 887333, 11235]
}
df = pd.DataFrame(data)

```

```python
# Melting the DataFrame
df_melted = pd.melt(df, id_vars=['anime', 'Completed', 'Dropped', 'Watching'], 
                     value_vars=['Score-1', 'Score-2', 'Score-3'],
                     var_name='Score_Type',
                     value_name='Score_Value')

# Create the line plot
sns.lineplot(x='anime', y='Score_Value', data=df_melted, hue='Score_Type', marker='o')

plt.title('Sample Line Plot')
plt.xlabel('Anime')
plt.ylabel('Score')

plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
df = pd.DataFrame(data)

# Create scatter plot
sns.scatterplot(x='Completed', y='Watching',data=df, hue='anime',s=100, alpha=0.8, marker= '^',palette="coolwarm")
plt.title("Scatter Plot of Completed, Dropped, and Watching for Each Anime")
plt.xlabel("Completed")
plt.ylabel("Dropped")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df = pd.DataFrame(data)

# Create bar plot
sns.barplot(x='anime', y='Completed',data=df, hue='anime',palette="rocket")
plt.title("Bar Plot of Completed Counts for Each Anime")
plt.yscale("log")  # Set y-axis to logarithmic scale
plt.ylabel("Completed")
plt.xlabel("Anime")
plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
# Plotting stacked bar chart
df.set_index('anime')[['Completed', 'Dropped', 'Watching']].plot(kind='bar', stacked=True, figsize=(10, 7), color=['#1f77b4', '#ff7f0e', '#2ca02c'])

plt.title("Stacked Bar Chart of Completed, Dropped, and Watching for Each Anime")
plt.xlabel("Anime")
plt.ylabel("Counts")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Status", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df_melted = df.melt(id_vars='anime', value_vars=['Score-1', 'Score-2', 'Score-3'], 
                    var_name='Score Type', value_name='Score')

# Create a box plot
sns.boxplot(y='Score Type', x='Score', data=df_melted, hue='Score Type', palette="viridis")
plt.title("Box Plot of Scores for Each Score Type")
plt.ylabel("Score")
plt.xlabel("Score Type")
plt.show()
```

```python
# Melt data for easier plotting
df_melted = df.melt(id_vars='anime', value_vars=['Score-1', 'Score-2', 'Score-3'], 
                    var_name='Score Type', value_name='Score')

# Create a violin plot
sns.violinplot(x='Score Type', y='Score', data=df_melted, hue='Score Type', palette="magma", density_norm= 'count')
plt.title("Violin Plot of Scores for Each Score Type")
plt.ylabel("Score")
plt.xlabel("Score Type")
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df[['Score-1', 'Score-2', 'Score-3', 'Completed', 'Dropped', 'Watching']].corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

```python
# Create a histogram for Score-1
plt.figure(figsize=(8, 6))
sns.histplot(df['Score-1'], bins=10, kde=True, color="skyblue", element= "poly")
plt.title("Histogram of Score-1")
plt.xlabel("Score-1")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create KDE plot for Score-1, Score-2, and Score-3
plt.figure(figsize=(10, 6))

sns.kdeplot(df['Score-1'], fill=True, color="skyblue", label="Score-1")
sns.kdeplot(df['Score-2'], fill=True, color="salmon", label="Score-2")
sns.kdeplot(df['Score-3'], fill=True, color="lightgreen", label="Score-3")

plt.title("KDE Plot for Score-1, Score-2, and Score-3")
plt.xlabel("Scores")
plt.ylabel("Density")
plt.legend(title="Scores")
plt.show()
```

```python
# Create a pair plot for Score-1, Score-2, and Score-3
sns.pairplot(df[['Score-1', 'Score-2', 'Score-3']])

# Show the plot
plt.suptitle("Pair Plot for Score-1, Score-2, and Score-3", y=1.02)
plt.show()
```

```python
# Create a joint plot for Score-1 and Score-2 combining scatter with KDE
sns.jointplot(x='Score-1', y='Score-2', data=df, kind='kde', marginal_ticks=True)

# Show the plot
plt.suptitle("Joint Plot Combining Scatter and KDE for Score-1 and Score-2", y=1.02)
plt.show()
```
