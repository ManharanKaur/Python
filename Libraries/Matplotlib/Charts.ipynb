```python
#Line Chart

#import important libraries
import matplotlib.pyplot as plt 
import numpy as np
# Define the x-coordinates (input data)
x_coordinates = np.array([0,1,2,3,4,5])
# Define the corresponding y-coordinates (output data)
y_coordinates = np.array([0,2,5,9,14,20])
# Plot the line graph using the x and y coordinates
plt.plot(x_coordinates, y_coordinates)
#Display the Line Chart
plt.show()
```

```python
#Line Graph (plotting without lines or simply scatter plot)

x_coordinates = np.array([0,1,2,3,4,5])
y_coordinated = np.array([0,2,6,10,18,24])

plt.plot(x_coordinates,y_coordinated, 'o')
plt.show()
```

```python
#Line graph(plotting with markers)
x_coordinates = np.array([0,1,2,3,4,5])
y_coordinated = np.array([0,2,6,10,18,24])

plt.plot(x_coordinates,y_coordinated, marker = 'o')
plt.show()
```

```python
#Pie Chart
# Create an array of data values
data = np.array([1,2,3,4,2,4,6,8,3,6,9,12,4,8,12,16])
# Create a pie chart from the data
plt.pie(data)
# Display the pie chart
plt.show()
```

```python
# Define the data for the x-axis (labels) and y-axis (frequencies)
x_axis = np.array(['data1','data2','data3','data4','data5'])
y_axis = np.array([24,54,36,47,29])

# Create the bar chart
plt.bar(x_axis, y_axis)

# Display the chart
plt.show()
```
