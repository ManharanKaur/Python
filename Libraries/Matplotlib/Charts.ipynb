```python
#Line Chart

#import important libraries
import matplotlib.pyplot as plt 
import numpy as np

# Line 1
# Define the x-coordinates (input data)
x_coordinates = np.array([0,1,2,3,4,5])
# Define the corresponding y-coordinates (output data)
y_coordinates = np.array([0,2,5,9,14,20])


# Plot the line graph using the x and y coordinates
plt.plot(x_coordinates,
         y_coordinates, 
         c = "#0CBC19", 
         lw = 3, 
         ls = '--', 
         marker = 'o', 
         ms = 10, 
         mec = 'blue', 
         mfc = 'cyan')


# Line 2
x_coordinates = np.array([1,2,3,4,5,6])
y_coordinates = np.array([6,5,4,3,2,1])
plt.plot(x_coordinates,
         y_coordinates, 
         c = "orange", 
         lw = 3, 
         ls = '-.', 
         marker = '*', 
         ms = 10, 
         mec = 'black', 
         mfc = '#A62BDB')


plt.xlabel('X axis')
plt.ylabel('y axis')
plt.grid(True)
plt.title('Line Chart')


#Display the Line Chart
plt.show()
```

```python
#Pie Chart

# Create an array of data values
data = np.array([2,3,4,2,4,6,8,3,6,9,12,4,8,12,16])
Lable = np.array(['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O'])
MyExplode = np.array([0,0,0.2,0,0,0,0,0,0,0.3,0,0,-0.1,0,0.5])

# Create a pie chart from the data
plt.pie(np.sort(data), 
        labels = Lable, 
        autopct = '%1.0f%%', 
        explode = MyExplode, 
        shadow = True, 
        startangle = 90, 
        radius = 1.3, 
        pctdistance = 0.9)
plt.title('Pie Chart')
plt.legend(title = 'Sample data')

# Display the pie chart
plt.show()
```

```python
# Bar Graph
# Define the data for the x-axis (labels) and y-axis (frequencies)
x_axis = np.array(['data1','data2','data3','data4','data5','data6','data7'])
y_axis = np.array([24,54,36,47,29,40,15])

# Create the bar chart
plt.bar(x_axis,
        y_axis, 
        width = 0.5, 
        color = '#F184B6', 
        edgecolor = 'blue')

plt.title('Bar Chart')
plt.xlabel('Data Name')
plt.ylabel('Data Values')
plt.grid(True)

# Display the chart
plt.show()
```

```python
# Histogram
# Creating random data the data
data = np.random.randint(low=10, high=81, size=100)
print(data)

#plotting the hinstgram
plt.hist(data,
        bins = 7,
        histtype = 'bar',
        color = '#A62BDB',
        edgecolor = 'black',
        alpha = 0.7,
        orientation = 'vertical',
        align = 'mid')

plt.title('Histogram')
plt.xlabel('Range')
plt.ylabel('Entries')
plt.grid(True)


#Displaying graph
plt.show()
```

```python
# scatter plot
# Creating random data for scatter plot
# plot 1
x_axis = np.random.randint(low = 0, high = 21, size = 50)
y_axis = np.random.randint(low = 0, high = 21, size = 50)
size = np.random.randint(low = 15, high = 500, size = 50)
# plotting the graph
plt.scatter(x_axis,
           y_axis,
           s = size,
           marker = 'o',
           edgecolor = 'orange')


# plot 2
x_axis = np.random.randint(low = 0, high = 21, size = 50)
y_axis = np.random.randint(low = 0, high = 21, size = 50)

plt.scatter(x_axis,
           y_axis,
           s = size,
           marker = '^',
           edgecolor = 'green')

# plot 3
x_axis = np.random.randint(low = 0, high = 21, size = 50)
y_axis = np.random.randint(low = 0, high = 21, size = 50)

plt.scatter(x_axis,
           y_axis,
           s = size,
           marker = '*')

plt.xlabel('X axis')
plt.ylabel('y axis')
plt.grid(True)
plt.title('Line Chart')

#Displaying the graph
plt.show()
```

```python
# Box Plot
# Creating random data for Box Plot
data = np.array([10,4,11,7,3,25,12,11,9,5,5,8])

#plotting the box plot
plt.boxplot(data,
           notch = True,
           vert = False,
           patch_artist = True,
           showmeans = True,
           boxprops = dict(facecolor='skyblue', color = 'red'),
           flierprops=dict(marker='h', markersize=10, markerfacecolor='red'))

plt.xlabel('Data series')
plt.grid(True)
plt.title('Box Plot')

#Displaying the graph
plt.show()
```

```python
# Area plot
# Creating random data for area plot
x = np.linspace(0, 10, 100)
print(x)
y1 = np.sin(x)
y2 = np.cos(x)

# Plotting the area
plt.subplot(2,1,1) # rows,columns, index
plt.fill_between(x, y1, y2,
                color = 'yellow',
                alpha = 0.4,
                # where = y1 > y2,
                hatch = '//',
                edgecolor = 'green')

plt.subplot(2,1,2) # rows,columns, index
plt.fill_between(x, y1, y2,
                color = 'yellow',
                alpha = 0.4,
                where = y1 > y2,
                hatch = '//',
                edgecolor = 'green')

plt.xlabel('X axis')
plt.ylabel('y axis')
plt.grid(True)
plt.title('Area Plot')

#Displaying the graph
plt.show()
```

```python
# Heat map
# Creating random matrix data for the heatmap
data = np.random.rand(5, 5)

# Create a heatmap
plt.subplot(3,3,1)
plt.imshow(data, 
           cmap='hot', 
           interpolation='bicubic',
           aspect = 'equal')

plt.subplot(3,3,2)
plt.imshow(data, 
           cmap='hot', 
           interpolation='bicubic',
           aspect = 'auto')

plt.subplot(3,3,3)
plt.imshow(data, 
           cmap='hot', 
           interpolation='nearest',
           aspect = 'auto')

plt.subplot(3,3,4)
plt.imshow(data, 
           cmap='hot', 
           interpolation='spline16')

plt.subplot(3,3,5)
plt.imshow(data, 
           cmap='hot', 
           interpolation='bilinear')

plt.subplot(3,3,6)
plt.imshow(data, 
           cmap='cool',
           aspect = 'auto')

plt.subplot(3,3,7)
plt.imshow(data, 
           cmap='Set1')

plt.subplot(3,3,8)
plt.imshow(data, 
           cmap='rainbow')

plt.subplot(3,3,9)
plt.imshow(data,
          cmap= 'DarPurplek2')

# Display the plot
plt.show()
```

```python
#Counter Plot
# Data for the contour plot
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a contour plot
plt.contour(X, Y, Z, levels=10, cmap='plasma')

# Add labels and title
plt.title('Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.colorbar()
plt.show()
```

```python
# Data for the stack plot
x = np.arange(10)
y1 = np.random.randint(1, 10, size=10)
y2 = np.random.randint(1, 10, size=10)
y3 = np.random.randint(1, 10, size=10)

# Create a stack plot
plt.stackplot(x, y1, y2, y3, labels=['Series 1', 'Series 2', 'Series 3'], alpha=0.8)

# Add labels and title
plt.title('Stack Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add legend
plt.legend(loc='upper left')

# Display the plot
plt.show()
```
